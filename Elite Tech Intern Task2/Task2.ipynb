```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("unemployment_post_covid.csv")
```

```python
# Display basic info
print("Dataset Info:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())
```

```python
df.dropna(inplace=True)  # Remove missing values
```

```python
# Plot Unemployment Rate Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Date", y="Unemployment_Rate", marker="o", hue="State")
plt.title("Unemployment Rate Over Time (Post-COVID)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.legend(title="State", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Bar Chart: Average Unemployment Rate by State
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="State", y="Unemployment_Rate", estimator=lambda x: sum(x)/len(x), hue="State", legend=False, palette="coolwarm")
plt.title("Average Unemployment Rate by State (2021-2024)")
plt.xlabel("State")
plt.ylabel("Avg. Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram to show the distribution of Unemployment Rate
plt.figure(figsize=(8, 5))
sns.histplot(df["Unemployment_Rate"], bins=20, kde=True, color="blue")
plt.title("Distribution of Unemployment Rate (Post-COVID)")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Save the processed dataset
df.to_csv("cleaned_unemployment_post_covid.csv", index=False)
print("Processed dataset saved as 'cleaned_unemployment_post_covid.csv'")
```
